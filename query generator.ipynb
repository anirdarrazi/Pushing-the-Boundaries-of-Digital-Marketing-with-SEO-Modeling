```python
from openai import OpenAI
import pandas as pd
import csv
```

```python
def get_last_word(s):
    # Split the string into words
    words = s.split()

    # Return the last word
    return words[-1] if words else ''
```

```python
def remove_last_word(s):
    # Split the string into words
    words = s.split()

    # Remove the last word
    words = words[:-1]

    # Join the remaining words back into a string
    return ' '.join(words)
```

```python
def export_list_to_csv(my_list, filename):
    try:
        with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            for item in my_list:
                writer.writerow([item])
        print(f"{len(my_list)} items exported to {filename}")
    except Exception as e:
        print(f"Error: {e}")
```

```python
def translate_to_language(language, prompt):
    # Initialize the OpenAI client
    client = OpenAI(api_key="...")

    # Create a chat completion request
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": f"You are a helpful assistant that translates English to {language}"},
            {"role": "user", "content": f"Translate this: {prompt}"}
        ]
    )

    # Access the assistant's reply content and return it
    assistant_reply = completion.choices[0].message.content
    #print(assistant_reply)
    return assistant_reply
```

```python
def process_csv_files():
    query_list = []
    keyword_list = []
    location_list = []
    language_list = []
    country_list = []

    try:
        # Read the first CSV file
        queries = pd.read_csv("C:/Users/Anir/iCloudDrive/Skola/Masteruppsats DSV VT24/Data/queries.csv")

        # Read the second CSV file
        cities = pd.read_csv("C:/Users/Anir/iCloudDrive/Skola/Masteruppsats DSV VT24/Data/cities.csv")
    
        temp_sub = 0
        # Loop through the first CSV file
        print("Starting the generation of queries...")
        for queries_index, queries_row in queries.iterrows():
            print("Main process: " + str(int(((queries_index + 1)/(len(queries) + 1))*100)) + "%")

            for cities_index, cities_row in cities.iterrows():

                if int(((cities_index + 1)/(len(cities) + 1))*100) == temp_sub:
                
                    print("Sub process: " + str(temp_sub) + "%")
                    temp_sub += 1
                #print("Sub process: " + str(int(((cities_index + 1)/(len(cities) + 1))*100)) + "%")
                if cities_row[2].strip() == "English":
                    original_string = queries_row[0]

                    # Define the substring to be replaced and the replacement string
                    substring_to_replace = "(Location English)"
                    replacement_string = cities_row[0]

                    # Use str.replace() to replace the substring
                    new_string = original_string.replace(substring_to_replace, replacement_string)

                    substring_to_replace_for_keyword = " (Location English)"
                    replacement_string_for_keyword = ""

                    # Use str.replace() to replace the substring
                    new_string_for_keyword = original_string.replace(substring_to_replace_for_keyword, replacement_string_for_keyword)
                
                    query_list.append(new_string)
                    
                    keyword_list.append(new_string_for_keyword)
                    location_list.append(cities_row[0])
                    language_list.append(cities_row[2])
                    country_list.append(cities_row[1])
                    #print("test")
                else:
                    original_string = queries_row[0]

                    substring_to_replace = "(Location English)"
                    replacement_string = cities_row[0]

                    # Use str.replace() to replace the substring
                    new_string = original_string.replace(substring_to_replace, replacement_string)
                    new_string_translation = translate_to_language(cities_row[2], new_string)
                    
                    #substring_to_replace_for_keyword = " (Location English)"
                    #replacement_string_for_keyword = ""

                    # Use str.replace() to replace the substring
                    #new_string_for_keyword = original_string.replace(substring_to_replace_for_keyword, replacement_string_for_keyword)
                    #print(new_string_translation)
                    query_list.append(new_string_translation)
                    
                    keyword_list.append(remove_last_word(new_string_translation))
                    location_list.append(get_last_word(new_string_translation))
                    language_list.append(cities_row[2])
                    country_list.append(cities_row[1])

    except Exception as e:
        print(f"Error: {e}")

    print("Done!")
    
    df = pd.DataFrame({
    'Queries': query_list,
    'Keyword': keyword_list,
    'Location': location_list,
    'language': language_list,
    'Country': country_list,
    
    })
    
    #export_list_to_csv(query_list, "eu_us_geo_industry_queries.csv")
    df.to_csv("all_queries.csv", index=True)
    return df
```
