```python
# Function for SERP data, API from DataForSEO

def seodata_serp1(queries_list):
    response_list = []

    for queries_index, queries in enumerate(queries_list["Queries"][:3]):
        # You can download this file from here https://cdn.dataforseo.com/v3/examples/python/python_Client.zip
        client = RestClient("...", "...")
        post_data = dict()
        post_data[len(post_data)] = dict(
        language_name= queries_list["language"][queries_index],
        location_name= queries_list["Country"][queries_index],
        keyword=queries,
        depth=100,
        )
        # POST /v3/serp/google/organic/live/advanced
        # in addition to 'google' and 'organic' you can also set other search engine and type parameters
        # the full list of possible parameters is available in documentation
        response = client.post("/v3/serp/google/organic/live/advanced", post_data)
        # you can find the full list of the response codes here https://docs.dataforseo.com/v3/appendix/errors
        if response["status_code"] == 20000:
            print("seodata_serp1" + " - " + queries + " - " + queries_list["language"][queries_index] + " - " + queries_list["Country"][queries_index] + " - " + str(queries_index) + " - " + str(int(((queries_index + 1)/(len(queries_list["Queries"][:3]) + 1))*100)) + "%")
            print(response)
            response_list.append(response)

        else:
            print("error. Code: %d Message: %s" % (response["status_code"], response["status_message"]))

    with open('serp_data1.json', 'w') as file:
        json.dump(response_list, file)
    
    print("seodata_serp1 done!")
    result1 = response_list
    #return response_list
```

```python
# Function for SERP data, API from DataForSEO

def seodata_serp(queries_list):
    response_list = []

    for queries_index, queries in enumerate(queries_list["Queries"]):
        # You can download this file from here https://cdn.dataforseo.com/v3/examples/python/python_Client.zip
        client = RestClient("...", "...")
        post_data = dict()
        post_data[len(post_data)] = dict(
        language_name= queries_list["language"][queries_index],
        location_name= queries_list["Country"][queries_index],
        keyword=queries,
        depth=100,
        )
        # POST /v3/serp/google/organic/live/advanced
        # in addition to 'google' and 'organic' you can also set other search engine and type parameters
        # the full list of possible parameters is available in documentation
        response = client.post("/v3/serp/google/organic/live/advanced", post_data)
        # you can find the full list of the response codes here https://docs.dataforseo.com/v3/appendix/errors
        if response["status_code"] == 20000:
            print(queries + " - " + queries_list["language"][queries_index] + " - " + queries_list["Country"][queries_index] + " - " + str(queries_index) + " - " + str(int(((queries_index + 1)/(len(queries_list["Queries"]) + 1))*100)) + "%")
            #print([queries_list["Keyword"][queries_index], queries_list["Location"][queries_index], response])
            response_list.append([queries_list["Keyword"][queries_index], queries_list["Location"][queries_index], queries_list["Country"][queries_index], response])

        else:
            print("error. Code: %d Message: %s" % (response["status_code"], response["status_message"]))
    
    print("Done!")
    return response_list
```
